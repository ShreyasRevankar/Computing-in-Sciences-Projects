```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_map(x, al):
    return al*x*(1-x)
```

```python
alpha = 3
x0 = 0.25
l1 = [x0]

for i in range(1, 10):
    l1.append(my_map(l1[-1], alpha))
    
l1
```

```python
def give_seq(x0, alpha, num_i):
    l1 = [x0]
    for i in range(1, num_i):
        l1.append(my_map(l1[-1], alpha))
    
    return l1
```

```python
plt.figure(figsize=(20,46))
x0 = 0.25
n_iter = 100
i = 0
for a in np.linspace(0.1, 4., 100):
    i+=1
    plt.subplot(20, 5, i)
    plt.plot(give_seq(x0, a, n_iter))
```

```python
def line_map(x, m, c=0):
    return m*x + c
```

```python
x_init = 1
```

```python
def line_seq(x0, num_i, m, c=0):
    l1 = [x0]
    for i in range(1, num_i):
        l1.append(line_map(l1[-1], m, c))
    
    return l1
```

```python
plt.plot(line_seq(x_init, 10, -0.5))
```

```python
plt.plot(line_seq(x_init, 10, -2))
```

```python
plt.plot(line_seq(x_init, 10, 2, 2))
```
