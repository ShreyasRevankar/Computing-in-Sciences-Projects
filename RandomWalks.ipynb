```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(num_steps, x_start=0):
    a_steps = np.random.randint(2, size=num_steps)
    b_steps = a_steps * 2 - 1
    return x_start + np.cumsum(b_steps)
```

```python
xi = np.zeros(100)
xi_sq = np.zeros(100)

all_pos =[]

for i in range(100):
    rw1 = random_walk(100)
    all_pos.append(rw1)
    xi += rw1
    xi_sq += (rw1*rw1)
    plt.plot(rw1)
    plt.xlabel('Step')
    plt.ylabel('Position')
```

```python
plt.plot(xi/100)
plt.xlabel('time')
plt.ylabel('<x_i>')
```

```python
plt.plot(xi_sq/100)
plt.xlabel('time')
plt.ylabel('<x_i^2>')
```

```python
plt.plot(xi_sq/100 - ((xi)/100)*((xi)/100))
plt.xlabel('x_i')
plt.ylabel('Variance')
```
