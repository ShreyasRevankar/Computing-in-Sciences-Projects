# Logistic Map

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

## Task 1

### Task 1.1

```python
def log_eq(x, alpha):
    '''
    Returns the value alpha*x(1-x)
    '''
    return alpha * x * (1-x)
```

```python
def log_map(x, alpha, n):
    '''
    Returns a list [x_0, x_1, x_2, ... , x_n] where x_r = x_r-1
    '''
    l1 = [x]
    for i in range(1, n):
        l1.append(log_eq(l1[-1], alpha))
    return l1
```

## Task 2

```python
def phase_plot(x, alpha, num_i):
    '''
    x: starting point
    alpha: parameter of the logistic equation
    num_i: number of iterations

    For given x, alpha, num_i it will plot the phase plot
    '''
    xvlas = log_map(x, alpha, num_i)
    yvals = log_map(log_eq(x, alpha), alpha, num_i)
    plt.plot(np.linspace(0, 1, 101) ,log_eq(np.linspace(0, 1, 101), alpha), label="y = f(x)")
    plt.plot(xvlas, yvals, label="trajectory")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
```

```python
phase_plot(0.5, 3.9, 100)
```

```python
def fnx(x, alpha):
    '''
    return f^15 (x)
    '''
    return log_map(x, alpha, 15)[-1]

def der_fnx(x, alpha):
    '''
    returns derivative of f^(15) (x)
    '''
    der = 1
    for x in log_map(x, alpha, 15):
        der *= alpha * (1 - 2 * x)
    return der

def stable_fixed_pts(alpha, tolr=0.005):
    '''
    alpha: parameter of logistic equation
    tolr: tolrence between fixed points (defaults to 0.005)
    '''
    fxpts = optimize.fixed_point(fnx, np.linspace(0.1, 0.75, 66), args=(alpha,))
    ans = []
    for pts in fxpts:
        if der_fnx(pts, alpha) <= 1:
            do_ap = True
            if len(ans) > 0:
                for p in ans:
                    if abs(p - pts) < tolr:
                        do_ap = False
            if do_ap:
                ans.append(pts)
    
    return ans
    # xvals = log_map(0.5, alpha, 32768)
    # der = 1
    # ans = []
    # for x in xvals:
    #     der *= alpha * (1 - 2 * x)

    # flg = False
    # # print(xvals[-10:])
    # for x in xvals[-128:-1]:
    #     if abs(x - xvals[-1]) < 0.00001:
    #         flg = True
    #         # print(x, xvals[-1])
    #         break
    # if der <= 1 and flg:
    #     my_x = xvals[-1]
    #     ans.append(my_x)
    #     for i in range(2000):
    #         my_x = log_eq(my_x, alpha)
    #         if abs(xvals[-1] - my_x) > tolr:
    #             ans.append(my_x)
    #         else:
    #             break
    # return ans
```
